```python
import pandas as pd

url_1 = 'https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv'
url_2 = 'https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv'

df_movies = pd.read_csv(url_1)
df_credits = pd.read_csv(url_2)

df_movies.head
```

```python
df_credits.columns
```

Dado que ambos dataframes tienen la columna 'title', los voy a unir con esta columna

```python
df = df_movies.merge(df_credits, on= 'title', how= 'inner')
```

```python
df.shape
```

```python
df = df[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
df.head()
```

```python
df.to_csv("movies.csv", index=False)
```

```python
df.duplicated().sum()
```

```python
type(df['genres'][0])
```

Parece que a pesar de tener forma de diccionario, los elementos de la columna en formato JSON están almacenados como strings, asique lo primero que haré será convertirlos en listas

```python
import json
df['genres'] = df['genres'].apply(lambda x: json.loads(x) if isinstance(x, str) else x)
df['name'] = df['genres'].apply(lambda x: [genre['name'] for genre in x]) #recorrer los diccionarios seleccionando 'name'

df = df.drop(['genres'], axis=1)
df.head()
```

```python
df['keywords'] = df['keywords'].apply(lambda x: json.loads(x) if isinstance(x, str) else x)
df['name_key'] = df['keywords'].apply(lambda x: [keyword['name'] for keyword in x]) 

df = df.drop(['keywords'], axis=1)
df.head()
```

```python
df['cast'] = df['cast'].apply(lambda x: json.loads(x) if isinstance(x, str) else x)
df['name_cast'] = df['cast'].apply(lambda x: [cas['name'] for cas in x[0:3]]) 

df = df.drop(['cast'], axis=1)
df.head()
```

```python
df['crew'][0]
```

```python
df['crew'] = df['crew'].apply(lambda x: json.loads(x) if isinstance(x, str) else x)
df['Director'] = df['crew'].apply(lambda x: [e['name'] for e in x if e['job']=='Director'])

df = df.drop(['crew'], axis=1)
df.head()
```

```python
df['overview'] = df['overview'].apply(lambda x: [x] if isinstance (x, str) else []) 
df.head()
```

```python
df[['name','name_cast', 'Director', 'name_key']] = df[['name','name_cast', 'Director', 'name_key']].map(
    lambda x: [s.strip().lower() for s in x]) #quitar el espacio en las columnas

df.head()
```

```python
df['tags'] = df[['name','name_cast', 'Director', 'name_key', 'overview']].apply(
    lambda row: ' '.join(row['overview'] + row['name'] + row['name_cast'] + row['name_key'] + row['Director']), axis=1)
#En las instrucciones se indica unir las filas con ',' y luego sustituirla por un espacio, pero directamente he puesto un espacio como separador
#Tampoco se indicaba un order
df.head()
```

# Vectorización de la data

En este caso como solo vamos a comparar datos dentro del dataframe no es necesario dividir en entrenamiento y test

```python
from sklearn.feature_extraction.text import CountVectorizer

vec_model = CountVectorizer(stop_words = "english")
vectors = vec_model.fit_transform(df['tags']).toarray()
```

```python
#Compare distance
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = df[df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse = True , key = lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(df.iloc[i[0]].title)
```

Vamos a probar si la función de recomendación funciona

```python
recommend("Avatar")
```
